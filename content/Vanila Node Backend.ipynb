# What are we trying to do?

Beginning of this course was inspired by [TODO], that aimed to teach you basics of frontend JS development, using just the in-browser javascript. No frameworks, no libraries.

While in later parts of the course we certainly won't shy away from using node package modules, we would first like to give you a taste of a vanilla nodejs experience. There is certain degree of abuse of the module system, where we have seen developers pull in packages such as [TODO, think it was first], that already have their native equivalent.

So, don't wory, we won't be reimplementing mongodb-client, or something similiar.

On the other hand, we will try to create something in decent scope, along the lines of

* lousy wget clone
* image store

After we are done, we will introduce npm ecosystem and

* discuss when to use libraries
* show how would we really write our pet projects
* how we would make an instalable module
